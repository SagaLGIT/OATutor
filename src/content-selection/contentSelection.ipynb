```python
import os
import numpy as np
import json
import pandas as pd
```

# Load files

```python
from collections import Counter

# json to matrix
with open('idLearningObj.json', 'r') as id_json:
    id_learning_objectives = json.load(id_json)

id_vals = list(id_learning_objectives.values())
flattened_id_vals = np.array([item for sublist in id_vals for item in sublist])
id_vals_counter = Counter(flattened_id_vals)
id_vals_counter
```

```python
json_lessons = pd.read_json('collegeAlgebraContent.json')
lessons = pd.DataFrame(json_lessons)
lessons
```

# Filter were 8 learning objectives are present

```python
lo_dict = lessons['learningObjectives']

# convert from json to list for each lesson
los = [list(lessons['learningObjectives'][i].keys()) for i in range(len(lo_dict))]

lessons['learningObjectives'] = los
lessons
```

```python
flattened_los = np.array([item for sublist in los for item in sublist])
unique_los = np.unique(flattened_los)

len(unique_los), len(flattened_los), unique_los
```

```python
T = 8
```

```python
over_T_lo = []
for lo in unique_los:
    n_lo = id_vals_counter[lo]
    # print(id_vals_counter[lo])
    if n_lo > T :
        print(lo, ':',n_lo)
        over_T_lo.append(lo)
```

# Connecting sufficient (learning objectives > 7) los with lessons

```python
lessons['Sufficient los'] = lessons['learningObjectives'].apply(lambda x: [x_i for x_i in x if x_i in over_T_lo])
lessons
```

# Inverting id to lo dict

```python
learning_objective_ids = {}


for  problem_id, learning_objectives in (id_learning_objectives.items()):
    for learning_objective in learning_objectives:
        if  learning_objective in learning_objective_ids.keys():
            learning_objective_ids[learning_objective].append(problem_id)
        else:
            learning_objective_ids[learning_objective] = [problem_id]

learning_objective_ids
```

# Randomly choosing from N lessons 

```python
N = 3
total_number_lessons = 9
```

```python
import random

final_selection = pd.DataFrame()

for i in range(1, total_number_lessons+1):
    lessons_i = lessons[lessons['name'].str.contains(f'Lesson {i}')]    
    rand_num = random.randint(1, len(lessons_i))
    chosen_lesson = lessons[lessons['name'].str.startswith(f'Lesson {i}.{rand_num}')]
    while chosen_lesson['Sufficient los'].values[0] == []:
        rand_num = random.randint(1, len(lessons_i))
        chosen_lesson = lessons[lessons['name'].str.startswith(f'Lesson {i}.{rand_num}')]

    final_selection = final_selection._append(chosen_lesson, ignore_index=True)

final_selection    
```

```python
# only taking first Learning Objective
final_learning_objectives = [row[0] for row in final_selection['Sufficient los']]
final_learning_objectives
```

```python
final_ids = [learning_objective_ids[lo] for lo in final_learning_objectives]
final_ids = [item for sublist in final_ids for item in sublist] # flatten
final_ids
```

# Write IDs to file

```python
id_out_file = 'selectedIDs.txt'
with open(id_out_file, 'w') as f:
    for id in final_ids:
        f.write("%s\n" % id)
```
